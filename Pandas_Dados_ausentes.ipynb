# Dados ausentes

Vamos mostrar alguns métodos convenientes para lidar com Missing Data em pandas:

```python
import numpy as np
import pandas as pd
```

```python
#Criando um DataFrame com dados ausentes

df = pd.DataFrame({'A':[1,2,np.nan],
                  'B':[5,np.nan,np.nan],
                  'C':[1,2,3]})
df
```

```python
#Retirando as colunas que tenham algum valor NaN
#Nao altera o DataFrame original

df.dropna()
```

```python
#Pode-se utilizar essa funcao escolhendo-se uma linha ou coluna

df.dropna(axis=1)
```

```python
#Tambem podemos utilizar a funcao thresh para selecionar as linhas
#especificando o numero de vezes que queremos que o NaN apareca para elimina-la

#Neste exemplo, apenas as linhas onde o NaN aparecer duas vezes serao eliminadas do DataFrame
df.dropna(thresh=2)
```

```python
#Podemos tambem inserir valores no DataFrame com i .fillna(valor= 'valor')
#que colocarar um valor no lugar de NAN.

df.fillna(value='Conteúdo')
```

```python
# Podemos ustilizar prara limpar os dados no DF de forma mais analítica
#Como exemplo inserir a media dos dados da coluna na propria coluna
#isto não altera a coluna, mas preenche do valor esperado para os dados


df['A'].fillna(value=df['A'].mean())
```
