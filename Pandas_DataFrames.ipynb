##Criação de um DataFrame, entradas e atributos.

```python
#CABECALHO E DEFINICOES INICIAIS


# DataFrame e um objeto da biblioteca Pandas

import pandas as pd
import numpy as np

#Neste notebook será utilizado uma matriz de valores aleatorios
#eles serao gerados pelo objeto rando da biblioteca numpy
# e sua semente necessita ser inicializada com o comando seed


from numpy.random import randn
np.random.seed(101)

#criando um DataFrame
#uma matriz 5x4
#cujas linhas sao os termos de 'A B C D'.split()
#cujas colunas sao os termos de 'W X Y Z'.split()
#(onde de forma esperta se usou o objeto split de strings)


df = pd.DataFrame(randn(5,4),index='A B C D E'.split(),columns='W X Y Z'.split())
df
```

## Seleção e indexação

Vamos aprender os vários métodos para pegar dados de um DataFrame

```python
#vamos passar o nome de uma coluna como argumento entre chaves para seleciona-la

df['W']
```

```python
# Passando uma lista com nomes das colunas
#teremos como resultado um DataFrame das duas colunas

df[['W','Z']]
```

** Criando uma coluna: **

    Basta se associar valores à coluna, como se estivessemos alterando os valores lá estocados.

```python
#A coluna new nao existe no DataFrame e sera criada com a asssicacao direta de valores
#nao e necessario o uso de argumentos como inplace

df['new'] = df['W'] + df['Y']
```

** Removendo colunas **

    O uso do comando drop
    DataFrame.drop('NomeColuna',axis= (0 ou 1))
    quando axis= 0 ---> retira uma linha
    quando axis= 1 ---> retira uma coluna

---

** ESTA EXCLUSÃO DE VALORES SÓ SERÁ INCORPORADA À MATRIZ ORIGINAL SE FIZERMOS USO DO PARÂMETRO: **

** inplace = True**

---



```python
#apagando coluna -- axis= 1
#nao altera o DataFrame original

df.drop('new',axis=1)
```

```python
# Porém, tal exclusão só ocorrerá se especificada no parâmetro inplace

df
```

```python
#utilizando-se o inplace, a alteracao e permanente
#cuidado!

df.drop('new',axis=1,inplace=True)
```

```python
df
```

Também podemos deletar linhas com o drop:
    
    axis= 0

```python
df.drop('E',axis=0)
```

** Selecionando linhas: **
    
    Função .loc['rótulo/label linha']['rótulo/label coluna']
      localiza por label ou rótulo das linhas ou colunas
    
    Função .iloc['rótulo/label linha']['rótulo/label coluna']
      localiza por índice da linha/entrada ou coluna/atributo


```python
df.loc['A']
```

```python
#loc para colunas e linhas
#um valor especifico

df.loc['A']['W']
```

Ou selecione com base na posição em vez do rótulo

```python
df.iloc[2]
```

** Selecionando o subconjunto de linhas e colunas **

```python
#uma forma mais pratica, emais utilizada
#e passar os argumentos como uma lista, 
#a primeira posicao esta relacionada a linha e a segunda a coluna

df.loc['B','Y']
```

```python
#pode se passar uma lista de colunas e linhas como argumento
#para se criar um novo DataFrame filtrado


df.loc[['A','B'],['W','Y']]
```

### Seleção condicional

Uma característica importante dos pandas é a seleção condicional usando notação de colchetes, muito semelhante ao numpy:

Aqui aparece o importante conceito de máscaras, sua criação a partir de sentença lógica, e o seu uso como um argumento de seleção do DataFrame.

```python
df
```

```python
#A mascara e uma matriz ou coluna que contem valores logicos
#Ela dira quais os valores serao acessados
#quando utilizada como argumento do DataFrame


#criando a mascara
df>0
```

```python
#acessando os valores onde a mascara tem valor True:

df[df>0]
```

```python
#Criando uma mascara para uma coluna
#ela selecionara as linhas onde o valor da mascara e True

df['W']>0
```

```python
#se a usarmos como argumento ele filtrara o DataFrame

df[df['W']>0]
```

```python
#O resultado da operacao acima e um DataFrame
#Podemos utilizar as mesmas ferramentas de selecao
#da mesma forma que utilizamos para um DataFrame

#selecionando uma coluna
df[df['W']>0]['Y']
```

```python
#selecionando duas colunas

df[df['W']>0][['Y','X']]
```

Para duas condições, você pode usar | e & com parênteses:

```python
df[(df['W']>0) & (df['Y'] > 1)]
```

## Hierarquia de índices e índices múltiplos

Vamos examinar como trabalhar com o Multi-Index, primeiro criaremos um exemplo rápido de como seria um DataFrame Multi-Indexado:

```python
# Níveis de Índice
outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside,inside))
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
hier_index
```

```python
df = pd.DataFrame(np.random.randn(6,2),index=hier_index,columns=['A','B'])
df
```

Agora vamos mostrar como indexar isso! Para a hierarquia de índice, usamos df.loc []. Se este fosse no eixo das colunas, você usaria a notação de suporte normal df []. Chamar um nível do índice retorna um sub-dataframe:

```python
df.loc['G1']
```

```python
df.loc['G1'].loc[1]
```

```python
df.index.names
```

```python
df.index.names = ['Grupo','Número']
```

```python
df
```

```python
df.xs('G1')
```

```python
df.xs(['G1',1])
```
